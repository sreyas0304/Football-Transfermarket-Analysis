```python
import pandas as pd
premier_league_data=pd.read_csv('premier-league-data.csv')
premier_league_data
```

```python
premier_league_data.info()
```

```python
premier_league_data.describe()
```

```python
premier_league_data['market_value'].fillna(premier_league_data['market_value'].mean(), inplace=True)
```

```python
premier_league_data
```

```python
premier_league_data['fee'].fillna(premier_league_data['fee'].mean(), inplace=True)
```

```python
premier_league_data
```

```python
premier_league_data['loan_status'].fillna("Not Applicable", inplace=True)
```

```python
premier_league_data
```

```python
premier_league_data['season'] = premier_league_data['season'].astype('category')
```

```python
premier_league_data
```

```python
premier_league_data.dtypes
```

```python
premier_league_data.drop_duplicates(inplace=True)
```

```python
premier_league_data
```

```python
Q1 = premier_league_data['market_value'].quantile(0.25)
Q3 = premier_league_data['market_value'].quantile(0.75)
IQR = Q3 - Q1
outliers_mask = (premier_league_data['market_value'] < (Q1 - 1.5 * IQR)) | (premier_league_data['market_value'] > (Q3 + 1.5 * IQR))
premier_league_data = premier_league_data[~outliers_mask]
```

```python
outliers_mask
```

```python
premier_league_data
```

```python
premier_league_data.reset_index(drop=True, inplace=True)
```

```python
premier_league_data
```

```python
premier_league_data['fee'] = premier_league_data['fee'].astype(int)
```

```python
premier_league_data
```

```python
premier_league_data.dtypes
```

```python
premier_league_data.isnull().sum()
```

```python
premier_league_data = premier_league_data.dropna(subset=['dealing_country'])
```

```python
premier_league_data
```

```python
premier_league_data = premier_league_data[premier_league_data['is_loan'] == False]
```

```python
premier_league_data
```

```python
premier_league_data.to_csv('premier-league-cleaned.csv',index=False)
```
